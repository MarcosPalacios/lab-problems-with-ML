## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    return r2_score(y_true, y_predict)
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
X = df_clean.drop('medv',axis=1)
y = df_clean['medv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42,shuffle=True)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Five separate RFR here with the given max depths

RFR = [RandomForestRegressor(max_depth=number) for number in [2,4,6,8,10]]
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths
for model in RFR:
    model.fit(X_train,y_train)
    
    y_predict_train = model.predict(X_train)
    y_predict_test = model.predict(X_test)
    
    print('The train score is', performance_metric(y_train,y_predict_train))
    print('The test score is',  performance_metric(y_test,y_predict_test))
    print('\n')
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?

### Bias-Variance Tradeoff
When the model is trained with a maximum depth of 1, does the model suffer from high bias or from high variance? How about when the model is trained with a maximum depth of 10?

```python
# Your response here

model = RandomForestRegressor(max_depth=1)
model.fit(X_train, y_train)
print('Max_depth = 1')
print(f'Test score: {r2_score(y_test, model.predict(X_test))}')
print(f'Train score: {r2_score(y_train, model.predict(X_train))}')
```

```python
# Your response here

model = RandomForestRegressor(max_depth=10)
model.fit(X_train, y_train)
print('Max_depth = 10')
print(f'Test score: {r2_score(y_test, model.predict(X_test))}')
print(f'Train score: {r2_score(y_train, model.predict(X_train))}')
```

```python
#Con depth = 1 los scores son bajos por lo que el bias mayor, no es preciso, 
#pero la varianza es menor por que la diferencia entre train y test es baja

#Pasa lo contrario con depth = 10 Los scores son altos por lo que el bias es menor, es mas preciso,
#pero la varianza es mayor al ser mayor la diferencia entre train y test.
```

### Best-Guess Optimal Model
What is the max_depth parameter that you think would optimize the model? Run your model and explain its performance.

```python
# Your response here

model = RandomForestRegressor(max_depth=8)
model.fit(X_train, y_train)
print('Max_depth = 10')
print(f'Test score: {r2_score(y_test, model.predict(X_test))}')
print(f'Train score: {r2_score(y_train, model.predict(X_train))}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth': np.linspace(0.5,10,20)}

model = GridSearchCV(RandomForestRegressor(), params)

model.fit(X_train, y_train)

model.best_score_,model.best_params_
```
